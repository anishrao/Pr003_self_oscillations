```python
'''
Get paths to corresponsing folders
'''

# get the path to the project's main directory (e.g., Project051)
BASE_DIR_PATH = os.path.dirname(os.path.abspath(os.getcwd()))

# ------------------------- data folder -----------------------

# get the path to the folder containing raw data
DATA_DIR_PATH = os.path.join(BASE_DIR_PATH, 'data','')

# get the path to the folder containing UV-Vis-NIR raw data
DATA_UV_DIR_PATH_UV = os.path.join(BASE_DIR_PATH, 'data', 'UV-Vis-NIR','2023-12-07-bspp-reversibility','uv','')
DATA_UV_DIR_PATH_QPOD = os.path.join(BASE_DIR_PATH, 'data', 'UV-Vis-NIR','2023-12-07-bspp-reversibility','')

# ------------------------- output folder -----------------------

# # get the path to the folder with generated figures
RESULT_DIR_PATH = os.path.join(BASE_DIR_PATH, 'results', 'figs', '')
```

```python
'''
Import uv data
'''

#================ import data =========================#

# Define the path to the folder containing the CSV files
folder_path = DATA_UV_DIR_PATH_UV

# Get a list of all CSV files in the folder
txt_files = glob.glob(os.path.join(folder_path, '*.txt'))
txt_files = natsorted(txt_files)
print(len(txt_files))
```

```python
#================ plot max abs data for reversibility cycle ===========================#
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(4, 2), dpi=300)

# Find max. abs.
max_abs = []

time = 0

for i, txt_file in enumerate(txt_files):
    # Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(txt_file, skiprows=14, delimiter='\t', header=None)
    new_df = df.iloc[432:1331, ] # 400 to 800 nm

    # Find the maximum value and its corresponding index
    del_abs = new_df[1].iloc[278] -  new_df[1].iloc[-1] #  abs at 526 nm - abs at 800 nm. This takes care of scattering
    
    time = time + 20 # 20 seconds
    
    # Plot the data
    axs.scatter(time/3600, del_abs, s=1, color = 'firebrick')  # time is in hours
    axs.set_xlabel('Time (h)')
    axs.set_ylabel('Extinction')
    axs.spines[['top', 'right']].set_visible(False)
    
    axs.set_xlim([0,2.5])
```

```python
#================ plot max abs data for reversibility cycle ===========================
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(4, 2), dpi=300)

def animate(frame):
    # Read the CSV file into a Pandas DataFrame
    txt_file = txt_files[frame]
    df = pd.read_csv(txt_file, skiprows=14, delimiter='\t', header=None)
    new_df = df.iloc[432:1331, ]  # 400 to 800 nm

    # Clear previous plot and plot the new data
    axs.clear()
    axs.plot(new_df[0], new_df[1], color='firebrick')
    axs.set_xlim([400,800])
    axs.set_ylim([0,0.45])
    axs.set_xlabel('Wavelength (nm)')
    axs.set_ylabel('Extinction')
    axs.spines[['top', 'right']].set_visible(False)

# Create the animation
animation = FuncAnimation(fig, animate, frames=len(txt_files), interval = 50, repeat = False)

# Show the plot
plt.show()
```

```python
import os

# Path to the folder containing the files
folder_path = DATA_UV_DIR_PATH_UV

# Get a list of all files in the folder
all_files = natsorted(os.listdir(folder_path))

# Set the initial file index
current_file_index = 21  # (since Python uses 0-based indexing)

# Set the step size for selecting files
step_size = 35

# Set the number of files you want to select
num_files_to_select = 501

# Initialize a list to store selected file names
selected_files = []

# Loop to select files based on the specified pattern
for _ in range(num_files_to_select):
    # Check if the current index is within the valid range
    if current_file_index < len(all_files):
        # Get the current file name
        current_file = all_files[current_file_index]

        # Add the current file to the list
        selected_files.append(current_file)

        # Update the current file index for the next iteration
        current_file_index += step_size
    else:
        # Break the loop if the index is out of range
        break

fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(4, 2), dpi=300)        
i = 1

for file_name in selected_files:
    print(file_name)
    df = pd.read_csv(DATA_UV_DIR_PATH_UV + file_name, skiprows=14, delimiter='\t', header=None)
    new_df = df.iloc[432:1331, ] # 400 to 800 nm

    # Find the maximum value and its corresponding index
    del_abs = new_df[1].iloc[278] -  new_df[1].iloc[-1] #  abs at 526 nm - abs at 800 nm. This takes care of scattering
    
    # Plot the data.
    axs.plot(i, del_abs)
    axs.scatter(i, del_abs, s=2, color = 'firebrick')
    axs.set_xlabel('Time (h)')
    axs.set_ylabel('Extinction')
    axs.spines[['top', 'right']].set_visible(False)
    
    i = i + 1


```
