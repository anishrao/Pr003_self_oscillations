# Calculating baseline

```python
baseline_fitter = Baseline(x_data=times_cool)
base_corr_abs = baseline_fitter.asls(max_values_cool, lam=1e7, p=0.02)
```

```python
def baseline_als(y, lam, p, niter=10):
  L = len(y)
  D = sparse.csc_matrix(np.diff(np.eye(L), 2))
  w = np.ones(L)
  for i in range(niter):
    W = sparse.spdiags(w, 0, L, L)
    Z = W + lam * D.dot(D.transpose())
    z = spsolve(Z, w*y)
    w = p * (y > z) + (1-p) * (y < z)
  return z
```

```python
def plot_fft(x, sr):
    """
    Plot the FFT (Fast Fourier Transform) of a signal.

    Parameters:
        x (array-like): Input signal.
        sr (float): Sampling rate (samples per second).

    Returns:
        None
    """
    # Number of samples
    N = len(x)
    # Time vector
    t = np.arange(0, N) / sr

    # Compute FFT
    X = np.fft.fft(x)
    # Frequency axis
    freq = np.fft.fftfreq(N, d=1/sr)

    # Plot FFT amplitude
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.scatter(freq, np.abs(X))
    plt.plot(freq, np.abs(X))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('|X(freq)|')
    plt.title('FFT Amplitude')
    plt.xlim(0,2)
    plt.yscale('log')

# Example usage
if __name__ == "__main__":
    # Sampling rate
    sr = 40  # Samples per second (1 hour duration)
    # Plot FFT
    plot_fft(base_corr[500:1000], sr)
```

## Performing fourier transform of the data

```python


def plot_fft(x, sr):
    """
    Plot the FFT (Fast Fourier Transform) of a signal.

    Parameters:
        x (array-like): Input signal.
        sr (float): Sampling rate (samples per second).

    Returns:
        None
    """
    # Number of samples
    N = len(x)
    # Time vector
    t = np.arange(0, N) / sr

    # Compute FFT
    X = np.fft.fft(x)
    # Frequency axis
    freq = np.fft.fftfreq(N, d=1/sr)

    # Plot FFT amplitude
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.stem(freq, np.abs(X))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('|X(freq)|')
    plt.title('FFT Amplitude')

    # Plot time domain signal
    plt.subplot(122)
    plt.plot(t, np.fft.ifft(X).real, 'r')  # Use only real part
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title('Time Domain Signal')
    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    # Sampling rate
    sr = 40  # Samples per second (1 hour duration)
    # Plot FFT
    plot_fft(max_values_cool, sr)
```
